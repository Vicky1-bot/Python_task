```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv(r"/content/sample_data/weather_data/DailyDelhiClimateTrain.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
#correlation matrix
# Calculate correlation matrix
selected_columns = ['meantemp', 'humidity', 'wind_speed']
correlation_matrix = data[selected_columns].corr()

# Create a heatmap plot
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, )
plt.title("Correlation Heatmap")
plt.show()
```

```python
#The date column in this dataset is not having a datetime data type.
#We will change it when required. Let’s have a look at the mean temperature in Delhi over the years:
```

```python
figure = px.line(data, x="date",
                 y="meantemp",
                 title='Mean Temperature in Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x="date",
                 y="humidity",
                 title='Humidity in Delhi Over the Years')
figure.show()
```

```python
# Now let’s have a look at the wind speed in Delhi over the years
```

```python
figure = px.line(data, x="date",
                 y="wind_speed",
                 title='Wind Speed in Delhi Over the Years')
figure.show()
```

```python
#Till 2015, the wind speed was higher during monsoons (August & September) and retreating monsoons (December & January).
# After 2015, there were no anomalies in wind speed during monsoons.
#Now let’s have a look at the relationship between temperature and humidity:
```

```python
figure = px.scatter(data_frame = data, x="humidity",
                    y="meantemp", size="meantemp",
                    trendline="ols",
                    title = "Relationship Between Temperature and Humidity")
figure.show()
```

```python
#There’s a negative correlation between temperature and humidity in Delhi.
#It means higher temperature results in low humidity and lower temperature results in high humidity.
```

```python
#Here’s how we can change the data type and extract year and month data from the date column:
```

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

```python
#Now let’s have a look at the temperature change in Delhi over the years
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

```python
#Although 2017 was not the hottest year in the summer, we can see a rise in the average temperature of Delhi every year.
```

```python
# The prophet model accepts time data named as “ds”, and labels as “y”. So let’s convert the data into this format:
```

```python
forecast_data = data.rename(columns = {"date": "ds",
                                       "meantemp": "y"})
print(forecast_data)
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
